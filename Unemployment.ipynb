```python
# Converting object datatype to categorical of different columns
Unemp['Region'] = Unemp['Region'].astype('category')
Unemp['Frequency'] = Unemp['Frequency'].astype('category')
Unemp['Area'] = Unemp['Area'].astype('category')
```

```python
# Checking the datatype after conversion
Unemp.dtypes
```

```python
# Plotting bar plot of Unemployment rate against Area
plt.figure(figsize=(14,4))
sns.barplot(x='Area', y='Unemployment_Rate', data=Unemp)
plt.title('Unemployment_Rate vs Area')
plt.xlabel('Area')
plt.ylabel('Unemployment_Rate')
plt.show()
```

```python
# Extract Year and Month from Date
Unemp['Year'] = pd.to_datetime(Unemp['Date']).dt.year
Unemp['Month'] = pd.to_datetime(Unemp['Date']).dt.month_name()

# Unemployment Category (Low, Medium, High)
bins = [0, 5, 15, 25, 77] # Define bins
labels = ['Low', 'Medium', 'High', 'Very High'] # Define labels
Unemp['Unemployment_Category'] = pd.cut(Unemp['Unemployment_Rate'], bins=bins, labels=labels, right=False) # create a new column

# Employment Growth Rate (Month-over-Month)
# Sort by Region, Area, and Date to ensure correct calculation
Unemp = Unemp.sort_values(by=['Region', 'Area', 'Date'])
# Calculate the percentage change in 'Employed' column
Unemp['Employment_Growth_Rate'] = Unemp.groupby(['Region', 'Area'])['Employed'].pct_change() * 100
# Fill NaN values with 0 if occured
Unemp['Employment_Growth_Rate'] = Unemp['Employment_Growth_Rate'].fillna(0)

# Flag data from April 2020 onwards as COVID-19 period
Unemp['COVID_19_Flag'] = Unemp['Date'].apply(lambda x: 'Yes' if x >= pd.to_datetime('2020-04-01') else 'No')

# Unemployment-to-Employment Ratio
Unemp['Unemployment_to_Employment_Ratio'] = Unemp['Unemployment_Rate'] / (100 - Unemp['Unemployment_Rate'])

# Seasonality Flag (Quarter)
def get_season(month):
    if month in [1, 2, 3]:
        return 'Q1'
    elif month in [4, 5, 6]:
        return 'Q2'
    elif month in [7, 8, 9]:
        return 'Q3'
    else:
        return 'Q4'

Unemp['Quarter_Season'] = Unemp['Month'].apply(get_season)

print("7 New features added successfully")
```

```python
# Checking the first 5 row of feature engineered dataset
Unemp.head()
```

```python
# Average unemployment rate before and after COVID-19
pre_covid = Unemp[Unemp['COVID_19_Flag'] == 'No']['Unemployment_Rate'].mean()
post_covid = Unemp[Unemp['COVID_19_Flag'] == 'Yes']['Unemployment_Rate'].mean()
print(f"Average Unemployment Rate (Pre-COVID): {pre_covid:.2f}%")
print(f"Average Unemployment Rate (Post-COVID): {post_covid:.2f}%")
```

```python
# Plotting bar plot of Unemployment rate against each Region
plt.figure(figsize=(14, 4))
sns.barplot(x='Region', y='Unemployment_Rate', data=Unemp, ci=None, palette='magma')
plt.xticks(rotation=90)
plt.title('Unemployment Rate by Region')
plt.ylabel('Unemployment_Rate')
plt.xlabel('Region')
plt.show()
```

```python
# Plot unemployment rate trends over time
plt.figure(figsize=(14, 5))
sns.lineplot(x='Date', y='Unemployment_Rate', data=Unemp, estimator='mean', ci=None, color='blue')
plt.axvline(pd.to_datetime('2020-04-01'), color='red', linestyle='--', label='COVID-19 Start')
plt.title('Unemployment Rate Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Creating new variable average metrics
avg_metrics = Unemp.groupby('Region')[['Unemployment_Rate', 
                                       'Labour_Participation_Rate']].mean()
```

```python
# Plotting pair bar plot of Percentage against the Labour participation rate by region
avg_metrics.sort_values(by='Unemployment_Rate', ascending=False).plot(kind='bar', figsize=(14, 4))
plt.title('Average Unemployment and Labor Participation Rate by Region')
plt.ylabel('Percentage')
plt.tight_layout()
plt.show()
```
